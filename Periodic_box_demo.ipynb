```python
import numpy as np
import matplotlib.pyplot as plt #changed
```

```python
def boxcheck (x, y):
    x_upper_limit = 10
    y_upper_limit = 10
    
    x_lower_limit = 0
    y_lower_limit = 0
    
    if x < x_upper_limit and x > x_lower_limit:
        if y < y_upper_limit and y > y_lower_limit:
            return np.array([x, y])
        if y > y_upper_limit:
            return np.array([x, y - y_upper_limit])
        if y < y_lower_limit:
            return np.array([x, y + y_lower_limit])
    if x > x_upper_limit:
        if ( y < y_upper_limit and y > y_lower_limit):
            return np.array([x- x_upper_limit, y])
        if y > y_upper_limit:
            return np.array([x- x_upper_limit, y - y_upper_limit])
        if y < y_lower_limit:
            return np.array([x- x_upper_limit, y + y_lower_limit])
    if x < x_lower_limit:
        if ( y < y_upper_limit and y > y_lower_limit):
            return np.array([x + x_lower_limit, y])
        if y > y_upper_limit:
            return np.array([x + x_lower_limit, y - y_upper_limit])
        if y < y_lower_limit:
            return np.array([x + x_lower_limit, y + y_lower_limit])
    
```

```python
h = 0.1

x = 1.0
y = 1.0

vx = -1.1
vy = 1.5

part = np.array([[x], [y], [vx], [vy]])
```

```python
timesteps = 200

for i in range(timesteps):
    x_new = part[0, -1] + vx*h
    y_new = part[1, -1] + vy*h
    position_check = boxcheck(x_new, y_new)
    part_new = np.array([[position_check[0]], [position_check[1]], [vx], [vy]])
    part = np.append(part, part_new, axis = 1)
```

```python
plt.scatter(part[0], part[1], s=0.5)
plt.title("Positions of particles")
plt.xlabel("x position")
plt.ylabel("y position")
#plt.xlim(0.0, 0.11)
#plt.ylim(0, 1)
plt.show()
```
